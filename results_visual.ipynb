```python
import numpy as np
import matplotlib.pyplot as plt

# Load the PCA eigenvalues
pca_eigenvalues = np.loadtxt('parallel_pca_eigenvalues.txt', skiprows=2)

# Plot
plt.figure(figsize=(10,6))
plt.plot(range(len(pca_eigenvalues)), pca_eigenvalues, 'o-', alpha=0.7)
plt.title('Eigenvalues of C matrix')
plt.xlabel('Index')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()

# Load the PCA singular values
pca_singular_values = np.loadtxt('parallel_pca_singular_values.txt', skiprows=2)

# Plot
plt.figure(figsize=(10,6))
plt.plot(range(len(pca_singular_values)), pca_singular_values, 'o-', alpha=0.7)
plt.title('Singular values of B matrix')
plt.xlabel('Index')
plt.ylabel('Singular value')
plt.grid()
plt.show()

# Load the PCA results
pca_results = np.loadtxt('parallel_pca_pca_results.txt')

# Plot
plt.figure(figsize=(10,6))
plt.scatter(pca_results[:,0], pca_results[:,1], alpha=0.5)
plt.title('First Two Principal Components')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid()
plt.show()
```
